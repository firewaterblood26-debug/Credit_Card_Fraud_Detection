```python
import pandas as pd

# Load the training dataset
train_df = pd.read_csv('/content/fraudtrain.csv')

# Load the testing dataset, skipping malformed lines
test_df = pd.read_csv('/content/fraudTest.csv', engine='python', on_bad_lines='skip')

print("Training Data Head:")
display(train_df.head())

print("\nTesting Data Head:")
display(test_df.head())
```

```python
print("Missing values in training data:")
missing_train = train_df.isnull().sum()
missing_train = missing_train[missing_train > 0]
missing_train_percent = (missing_train / len(train_df)) * 100
missing_train_df = pd.DataFrame({
    'Missing Count': missing_train,
    'Missing Percentage': missing_train_percent
}).sort_values(by='Missing Percentage', ascending=False)
display(missing_train_df)

print("\nMissing values in testing data:")
missing_test = test_df.isnull().sum()
missing_test = missing_test[missing_test > 0]
missing_test_percent = (missing_test / len(test_df)) * 100
missing_test_df = pd.DataFrame({
    'Missing Count': missing_test,
    'Missing Percentage': missing_test_percent
}).sort_values(by='Missing Percentage', ascending=False)
display(missing_test_df)
```

```python
print(f"Shape of train_df before dropping missing values: {train_df.shape}")
train_df.dropna(inplace=True)
print(f"Shape of train_df after dropping missing values: {train_df.shape}")

print(f"\nShape of test_df before dropping missing values: {test_df.shape}")
test_df.dropna(inplace=True)
print(f"Shape of test_df after dropping missing values: {test_df.shape}")
```

```python
print('Converting trans_date_trans_time to datetime objects...')
train_df['trans_date_trans_time'] = pd.to_datetime(train_df['trans_date_trans_time'])
test_df['trans_date_trans_time'] = pd.to_datetime(test_df['trans_date_trans_time'])

print('\nData types after conversion for train_df:')
display(train_df[['trans_date_trans_time']].info())

print('\nData types after conversion for test_df:')
display(test_df[['trans_date_trans_time']].info())
```

### Feature Engineering: Extracting Time-Based Features

```python
print('Extracting time-based features from trans_date_trans_time...')

def extract_time_features(df):
    df['year'] = df['trans_date_trans_time'].dt.year
    df['month'] = df['trans_date_trans_time'].dt.month
    df['day'] = df['trans_date_trans_time'].dt.day
    df['hour'] = df['trans_date_trans_time'].dt.hour
    df['minute'] = df['trans_date_trans_time'].dt.minute
    df['day_of_week'] = df['trans_date_trans_time'].dt.dayofweek
    df['week_of_year'] = df['trans_date_trans_time'].dt.isocalendar().week.astype(int)
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    return df

train_df = extract_time_features(train_df)
test_df = extract_time_features(test_df)

print('\nFirst 5 rows of train_df with new time features:')
display(train_df.head())

print('\nFirst 5 rows of test_df with new time features:')
display(test_df.head())
```

### Data Preprocessing: Handling Categorical Features

```python
print('Identifying and encoding categorical features...')

# Identify categorical columns (excluding the target variable 'is_fraud')
categorical_cols = [
    col for col in train_df.select_dtypes(include='object').columns
    if col != 'is_fraud'
]

print(f"Categorical columns to encode: {categorical_cols}")

# Apply one-hot encoding
train_df = pd.get_dummies(train_df, columns=categorical_cols, drop_first=True)
test_df = pd.get_dummies(test_df, columns=categorical_cols, drop_first=True)

print(f"Shape of train_df after one-hot encoding: {train_df.shape}")
print(f"Shape of test_df after one-hot encoding: {test_df.shape}")

print('\nTrain_df head after one-hot encoding:')
display(train_df.head())

print('\nTest_df head after one-hot encoding:')
display(test_df.head())
```

### Aligning Columns After One-Hot Encoding

```python
print('Aligning columns between train_df and test_df...')

# Get all columns from both dataframes
train_cols = set(train_df.columns)
test_cols = set(test_df.columns)

# Identify columns unique to train_df and test_df
missing_in_test = list(train_cols - test_cols)
missing_in_train = list(test_cols - train_cols)

# Add missing columns to test_df and fill with 0
for col in missing_in_test:
    test_df[col] = 0

# Add missing columns to train_df and fill with 0
for col in missing_in_train:
    train_df[col] = 0

# Ensure the order of columns is the same
test_df = test_df[train_df.columns]

print(f"Shape of train_df after alignment: {train_df.shape}")
print(f"Shape of test_df after alignment: {test_df.shape}")

print('\nTrain_df head after alignment:')
display(train_df.head())

print('\nTest_df head after alignment:')
display(test_df.head())
```

### Data Preprocessing: Dropping Irrelevant Columns

```python
print('Dropping irrelevant columns...')

# Columns to drop based on being identifiers or already processed
columns_to_drop = [
    'Unnamed: 0', 'trans_date_trans_time', 'cc_num', 'first', 'last',
    'street', 'trans_num', 'unix_time', 'dob'
]

train_df.drop(columns=columns_to_drop, inplace=True)
test_df.drop(columns=columns_to_drop, inplace=True)

print(f"Shape of train_df after dropping columns: {train_df.shape}")
print(f"Shape of test_df after dropping columns: {test_df.shape}")

print('\nTrain_df head after dropping columns:')
display(train_df.head())

print('\nTest_df head after dropping columns:')
display(test_df.head())
```

### Data Preprocessing: Scaling Numerical Features

```python
from sklearn.preprocessing import StandardScaler

print('Scaling numerical features...')

# Identify numerical columns (excluding the target variable 'is_fraud' and boolean columns from one-hot encoding)
numerical_cols = train_df.select_dtypes(include=['int64', 'float64']).columns.tolist()
if 'is_fraud' in numerical_cols:
    numerical_cols.remove('is_fraud')

# Initialize StandardScaler
scaler = StandardScaler()

# Apply scaling to numerical columns in both train_df and test_df
train_df[numerical_cols] = scaler.fit_transform(train_df[numerical_cols])
test_df[numerical_cols] = scaler.transform(test_df[numerical_cols])

print('\nFirst 5 rows of train_df after scaling numerical features:')
display(train_df.head())

print('\nFirst 5 rows of test_df after scaling numerical features:')
display(test_df.head())
```

### Preparing Data for Modeling: Splitting Features and Target

```python
print('Splitting data into features (X) and target (y)...')

X_train = train_df.drop('is_fraud', axis=1)
y_train = train_df['is_fraud']

X_test = test_df.drop('is_fraud', axis=1)
y_test = test_df['is_fraud']

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_test: {y_test.shape}")

print('\nFirst 5 rows of X_train:')
display(X_train.head())

print('\nFirst 5 rows of y_train:')
display(y_train.head())
```

```python
print('Initializing and training Logistic Regression model...')

from sklearn.linear_model import LogisticRegression

# Initialize Logistic Regression model with class_weight='balanced' to handle imbalance
log_reg = LogisticRegression(class_weight='balanced', solver='liblinear', random_state=42)

# Train the model
log_reg.fit(X_train, y_train)

print('Logistic Regression model trained successfully.')
```

```python
print('Evaluating Logistic Regression model performance...')

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report

# Make predictions on the test set
y_pred = log_reg.predict(X_test)
y_pred_proba = log_reg.predict_proba(X_test)[:, 1]  # Probabilities for the positive class

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

print(f"\nAccuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"ROC AUC: {roc_auc:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
print('Initializing and training Random Forest model...')

from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest classifier with class_weight='balanced'
# n_estimators can be adjusted, but 100 is a good starting point
# random_state for reproducibility
rf_model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42, n_jobs=-1)

# Train the model
rf_model.fit(X_train, y_train)

print('Random Forest model trained successfully.')
```

```python
print('Evaluating Random Forest model performance...')

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report

# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)
y_pred_proba_rf = rf_model.predict_proba(X_test)[:, 1]  # Probabilities for the positive class

# Calculate evaluation metrics
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)
roc_auc_rf = roc_auc_score(y_test, y_pred_proba_rf)

print(f"\nRandom Forest Accuracy: {accuracy_rf:.4f}")
print(f"Random Forest Precision: {precision_rf:.4f}")
print(f"Random Forest Recall: {recall_rf:.4f}")
print(f"Random Forest F1-Score: {f1_rf:.4f}")
print(f"Random Forest ROC AUC: {roc_auc_rf:.4f}")

print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))

print("\nRandom Forest Confusion Matrix:")
cm_rf = confusion_matrix(y_test, y_pred_rf)
print(cm_rf)
```

```python
print('Adjusting Random Forest classification threshold and re-evaluating...')

# Define the new threshold
new_threshold = 0.2

# Apply the new threshold to the predicted probabilities
y_pred_rf_threshold = (y_pred_proba_rf >= new_threshold).astype(int)

# Calculate evaluation metrics with the new threshold
accuracy_rf_thresh = accuracy_score(y_test, y_pred_rf_threshold)
precision_rf_thresh = precision_score(y_test, y_pred_rf_threshold)
recall_rf_thresh = recall_score(y_test, y_pred_rf_threshold)
f1_rf_thresh = f1_score(y_test, y_pred_rf_threshold)

# ROC AUC does not change with threshold adjustment, as it uses probabilities directly
roc_auc_rf_thresh = roc_auc_score(y_test, y_pred_proba_rf)

print(f"\nRandom Forest Accuracy (Threshold={new_threshold}): {accuracy_rf_thresh:.4f}")
print(f"Random Forest Precision (Threshold={new_threshold}): {precision_rf_thresh:.4f}")
print(f"Random Forest Recall (Threshold={new_threshold}): {recall_rf_thresh:.4f}")
print(f"Random Forest F1-Score (Threshold={new_threshold}): {f1_rf_thresh:.4f}")
print(f"Random Forest ROC AUC (Threshold={new_threshold}): {roc_auc_rf_thresh:.4f}")

print("\nRandom Forest Classification Report (Adjusted Threshold):")
print(classification_report(y_test, y_pred_rf_threshold))

print("\nRandom Forest Confusion Matrix (Adjusted Threshold):")
cm_rf_thresh = confusion_matrix(y_test, y_pred_rf_threshold)
print(cm_rf_thresh)
```

### Hyperparameter Tuning for Random Forest using GridSearchCV

```python
print('Starting GridSearchCV for Random Forest hyperparameter tuning...')

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Define the parameter grid to search
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [10, 20, None], # None means nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'class_weight': ['balanced'] # Keep balanced for imbalanced dataset
}

# Initialize a Random Forest Classifier
rf_base = RandomForestClassifier(random_state=42, n_jobs=-1)

# Initialize GridSearchCV
# Scoring 'recall' is chosen to prioritize identifying more fraudulent transactions
grid_search = GridSearchCV(
    estimator=rf_base,
    param_grid=param_grid,
    scoring='recall',
    cv=3, # Using 3-fold cross-validation for faster execution, can be increased for more robust results
    verbose=2,
    n_jobs=-1
)

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

print('\nGridSearchCV completed.')
print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best recall score found: {grid_search.best_score_:.4f}")
```

```python
print('Evaluating the best Random Forest model found by GridSearchCV...')

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report

# Get the best model from GridSearchCV
best_rf_model = grid_search.best_estimator_

# Make predictions on the test set using the best model
y_pred_best_rf = best_rf_model.predict(X_test)
y_pred_proba_best_rf = best_rf_model.predict_proba(X_test)[:, 1] # Probabilities for the positive class

# Calculate evaluation metrics
accuracy_best_rf = accuracy_score(y_test, y_pred_best_rf)
precision_best_rf = precision_score(y_test, y_pred_best_rf)
recall_best_rf = recall_score(y_test, y_pred_best_rf)
f1_best_rf = f1_score(y_test, y_pred_best_rf)
roc_auc_best_rf = roc_auc_score(y_test, y_pred_proba_best_rf)

print(f"\nBest Random Forest Accuracy: {accuracy_best_rf:.4f}")
print(f"Best Random Forest Precision: {precision_best_rf:.4f}")
print(f"Best Random Forest Recall: {recall_best_rf:.4f}")
print(f"Best Random Forest F1-Score: {f1_best_rf:.4f}")
print(f"Best Random Forest ROC AUC: {roc_auc_best_rf:.4f}")

print("\nBest Random Forest Classification Report:")
print(classification_report(y_test, y_pred_best_rf))

print("\nBest Random Forest Confusion Matrix:")
cm_best_rf = confusion_matrix(y_test, y_pred_best_rf)
print(cm_best_rf)
```
